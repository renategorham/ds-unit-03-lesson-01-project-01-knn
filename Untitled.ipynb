# k-Nearest Neighbor by hand

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
from sklearn.neighbors import KNeighborsClassifier
import lib.train_test as tt
import lib.my_knn as knn
import lib.optimal_k as optimal_k
%matplotlib inline
```

##### Data Set

```python
#read data set. From wisconson Breast Cancer
df = pd.read_csv('data/wdbc.csv')
```

##### Select two variables for kNN evaluation

```python
#variables are checked haphazardly, visualy observed for good separation.
#after a few observations, mean cell radius and mean concave points will be used.
plt.figure(figsize=(15,5))
#histogram of mean radius
plt.subplot(121)
plt.hist(df[df['diagnosis']=='M'].mean_radius,color=(0.1, 0.2, 0.5, 0.3),\
        label='Malignant')
plt.hist(df[df['diagnosis']=='B'].mean_radius,color=(0.3, 0.2, 0.5, 0.3),\
        label='Benign')
plt.title('Histogram of Mean Cell Radius\nby Diagnosis')
plt.xlabel('Mean Cell radius')
plt.ylabel('Count')
plt.legend()
plt.grid()

#histogram of mean concave points
plt.subplot(122)
plt.hist(df[df['diagnosis']=='M'].mean_concave_points,color=(0.1, 0.2, 0.5, 0.3),\
        label='Malignant')
plt.hist(df[df['diagnosis']=='B'].mean_concave_points,color=(0.3, 0.2, 0.5, 0.3),\
        label='Benign')
plt.title('Histogram of Mean Concave Points\nby Diagnosis')
plt.xlabel('Mean Cell radius')
plt.ylabel('Count')
plt.legend()
plt.grid()
plt.show()
```

##### Subset dataframe for training 

```python
df = df[['diagnosis', 'mean_radius', 'mean_concave_points']]
```

```python
wdbc_training = tt.train_test(df,'diagnosis')[0]
wdbc_test = tt.train_test(df,'diagnosis')[1]
```

##### Do k-Nearest Neighbor

```python
#scatter plot mean cell radius and mean concave points by diagnosis class
plt.scatter(wdbc_training[wdbc_training['diagnosis']=='M'].mean_radius,\
           wdbc_training[wdbc_training['diagnosis']=='M'].mean_concave_points,\
           color='g',label='Malignant')
plt.scatter(wdbc_training[wdbc_training['diagnosis']=='B'].mean_radius,\
           wdbc_training[wdbc_training['diagnosis']=='B'].mean_concave_points,\
           color='b',label='Benign')
plt.title('Scatter Plot of Mean Radius and\nMean Concave Points\nby Diagnosis')
plt.xlabel('Mean Radius')
plt.ylabel('Mean Concave Points')
plt.legend()
plt.grid()
plt.show()
```

##### Results for by hand method

```python
knn.kNN(5, wdbc_training, 'diagnosis', 'mean_radius', 'mean_concave_points', \
        wdbc_test, 'mean_radius', 'mean_concave_points', \
        predict_class='diagnosis')
```

##### Comparison with SKlearn

```python
neighbors = KNeighborsClassifier(n_neighbors=5)
X = wdbc_training[['mean_radius', 'mean_concave_points']]
Y = wdbc_training.diagnosis
neighbors.fit(X,Y)


neighbors.predict(wdbc_test[['mean_radius', 'mean_concave_points']])

count_class =  np.unique(neighbors.predict(wdbc_test[['mean_radius', 'mean_concave_points']]),\
                        return_counts=True)
count_class
```
